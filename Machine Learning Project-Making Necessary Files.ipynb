```python
import pandas as pd
import numpy as np
```

```python
main_data_frame=pd.read_csv('reduced_book_rating_hoang.csv')
```

```python
main_data_frame['pointer']=1
```

```python
split = main_data_frame[['isbn','pointer']] 
split['total_title_review']= split.groupby(split['isbn']).transform('sum')
split.drop_duplicates(['isbn'], inplace=True)
split.drop(['pointer'], axis=1, inplace=True)
main_data_frame=pd.merge(main_data_frame, split, on='isbn', how='left')
```

```python
split = main_data_frame[['isbn', 'rating']]
split['rating'] = split.rating.astype(float)
split['total_title_rating'] = split.groupby(split['isbn']).transform('sum')
split.drop_duplicates(['isbn'], inplace=True)
split.drop(['rating'], axis=1, inplace=True)
main_data_frame=pd.merge(main_data_frame, split, on='isbn', how='left')
```

```python
def average(cols):
    return float (cols[0])/float(cols[1])
main_data_frame['title_rating_average'] = main_data_frame[['total_title_rating','total_title_review']].apply(average,axis=1)
main_data_frame.head()
```

```python
main_data_frame['total_title_review'] = main_data_frame.total_title_review.astype(float)
main_data_frame['total_title_review']= pd.to_numeric(main_data_frame['total_title_review'])
book_with_review_rating=main_data_frame.copy()
book_with_review_rating=book_with_review_rating.loc[book_with_review_rating['rating']>6]
book_with_review_rating=book_with_review_rating.loc[book_with_review_rating['total_title_review']>=5]
```

```python
book_with_review_rating.head()
```

```python

book_with_review_rating.drop(['rating','pointer','age','country','total_title_rating','total_title_review','title_rating_average'], axis=1, inplace=True)
```

```python
book_with_review_rating.drop_duplicates(['isbn'], inplace=True)
```

```python
book_with_review_rating.shape
```

```python
book_with_review_rating.to_csv('/Users/hhn/Downloads/book_features_implicit.csv')
```

```python
split = main_data_frame[['user_id','pointer']] 
split['total_user_review']= split.groupby(split['user_id']).transform('sum')
split.drop_duplicates(['user_id'], inplace=True)
split.drop(['pointer'], axis=1, inplace=True)
main_data_frame=pd.merge(main_data_frame, split, on='user_id', how='left')
```

```python
split = main_data_frame[['user_id', 'rating']]
split['rating'] = split.rating.astype(float)
split['total_user_rating'] = split.groupby(split['user_id']).transform('sum')
split.drop_duplicates(['user_id'], inplace=True)
split.drop(['rating'], axis=1, inplace=True)
main_data_frame=pd.merge(main_data_frame, split, on='user_id', how='left')
main_data_frame
```

```python
main_data_frame['user_rating_average'] = main_data_frame[['total_user_rating_x','total_user_review_x']].apply(average,axis=1)
main_data_frame['total_user_review_x'] = main_data_frame.total_user_review_x.astype(float)
main_data_frame['total_user_review_x']= pd.to_numeric(main_data_frame['total_user_review_x'])




```

```python
users_attributes=main_data_frame.copy()
users_attributes=users_attributes.loc[users_attributes['rating']>6]
users_attributes=users_attributes.loc[users_attributes['total_user_review_x']>=10]
users_attributes.head()
```

```python

users_attributes.drop(['rating','title','author','oldest year_publication','isbn','pointer','total_title_rating','total_title_review','title_rating_average','total_user_review_x','total_user_rating_x','user_rating_average','total_user_review_y','total_user_rating_y'], axis=1, inplace=True)
users_attributes.head()
```
